## Carga de datos

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
diamonds = sns.load_dataset('diamonds')
```

Verificar si el dataframe diamonds tiene datos faltantes, caso existieran eliminar las filas que contengan datos faltantes.

```python
# Utilizando isnull() verificamos si se encuentra algún dato faltante dentro del dataset como valor Booleano 
# sumamos esos faltantes con la función sum, cada vez que encuentre un 'True' lo interpretará como 1 y sumará en cada columna
# podemos observar que ninguna de las columnas del dataset tiene datos faltantes, así que no es necesario eliminar las filas. 
diamonds.isnull().sum()
```

Mostrar las siguientes estadisticas para todas las columnas númericas del dataframe diamonds:

* media (mean)
* desviación estandar (std)
* valor minimo (min)
* valor máximo (max)
* Percentil 25%
* Percentil 50%
* Percentil 75%


```python
statistics = pd.DataFrame(diamonds[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']].describe(percentiles=[0.25, 0.5, 0.75]))
statistics

# utilizaremos la función describe para hallar los estadísticos deseados, para ello le pasaremos una lista de columnas numéricas que deseamos
# luego, al no incluir los percentiles, tendremos que pasar como parámetro a la función una lista con los percentiles deseados 25, 50, 75
# finalmente para darle una mejor visualización, usaremos pd.DataFrame para darle forma de dataframe a nuestros estadísticos.
```

Mostrar diagrama de boxplot para la columna "price" agrupada por "clarity"

```python
sns.boxplot(y="price", hue="clarity",data=diamonds,palette ="husl")
plt.title('Boxplot Diamonds Price grouped by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Price')
plt.legend(loc='center right', bbox_to_anchor=(1.2, 0.5))
plt.show()
# utilizamos la función sns.boxplot para generar el diagrama de cajas del precio agrupado por la claridad de cada uno de los diamantes
# utlizamos plt.title,xlabel,ylabel para poner un título, y etiquetas a los ejes
# utilizamos legend para poder ponerla fuera del gráfico con bbox_to_anchor y tener mejor visualización de la data
# Podemos observar que si bien se encuentran cajas con una variabilidad baja que va incrementando de izquierda a derecha, 
# existen datos atípicos, lo cual nos lleva a deducir que no es la claridad el factor que determina el precio
# de ello se desprende que es otra la variable que les da el valor
```

Mostrar diagramas de dispersión para todos los pares de columnas númericas en el dataframe y agrupar por "precio".

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))
# utilizamos fig, axes para poder asignar a cada uno de los plot el plot que deseemos en la posición que deseemos
# creamos un array bidimensional de plots de 2 filas y 3 columnas con nrows y ncols
# utilizamos figsize para poder tener un gráfico más ancho y largo, ello para tener la capacidad de poner mostrar gráficos con mayor detalle
sns.scatterplot(data=diamonds, x='price', y='carat', ax=axes[0, 0])
sns.scatterplot(data=diamonds, x='price', y='depth', ax=axes[0, 1])
sns.scatterplot(data=diamonds, x='price', y='table', ax=axes[0, 2])
sns.scatterplot(data=diamonds, x='price', y='x', ax=axes[1, 0])
sns.scatterplot(data=diamonds, x='price', y='y', ax=axes[1, 1])
sns.scatterplot(data=diamonds, x='price', y='z', ax=axes[1, 2])
# utilizamos la función sns.scatterplot para hacer el diagrama de dispersión de cada una de las variables con respecto al precio. 
# utilizamos ax=axes[] para indicarle al plot donde ha de estar posicionado
plt.tight_layout()
# utilizamos tight_layout para poder ajustar automaticamente los elementos y que no se superpongan 
plt.show()
```
